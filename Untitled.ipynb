```python
import re
import pprint

import hashlib
import json
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML
display(HTML("<style>.container { width:80% !important; }</style>"))
```

```python
def get_similarity_scores_by_id(data_list, target_id):
    for item in data_list:
        if item['ID'] == target_id:
            # Extract and return the two specific values
            return item['question_context_similarity']
    return None     # Return None for both if no match is found

def load_json_data(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    return data

# Load the CSV file
def load_csv(file_path):
    return pd.read_csv(file_path)
    
def filter_answers(df):
    # Use a regular expression to filter rows
    # The regex '^[A-Z]$' matches a single uppercase letter
    return df[df['Answer'].str.match(r'^[A-Z]$', na=False)]

file_path = 'NEJM_All_Questions_And_Answers.csv'  
df = load_csv(file_path)
df = filter_answers(df)

def load_results(filename):
    """Load results from a file into pandas DataFrame."""
    data_c = {}
    data_nc = {}
    embedding_distance = load_json_data('updated_results.json')
    # Assuming 'df' and 'get_similarity_scores_by_id' are defined elsewhere and accessible here
    df.set_index('ID', inplace=True)  # Ensure 'df' is set correctly with 'ID' as index

    with open(filename, 'r') as f:
        for line in f:
            v_t = json.loads(line)
            for v in v_t:
                c_len = len(v['context'])
                answer = v['response'][0]
                correctness = (answer == df.loc[int(v['id']), 'Answer'])
                
                if c_len == 0:
                    data_nc[v['id']] = {
                        'id': v['id'],
                        'answer':answer,
                        'correctness': correctness,
                    }
                else:
                    context_similarity = get_similarity_scores_by_id(embedding_distance, int(v['id']))
                    data_c[v['id']] = {
                        'id': v['id'],
                        'context_length': c_len,
                        'answer':answer,
                        'c_answer':df.loc[int(v['id']), 'Answer'],
                        'correctness': correctness,
                        'similarity': context_similarity,
                    }

    # Transforming both dictionaries into pandas DataFrames
    data_c = pd.DataFrame.from_dict(data_c, orient='index')
    data_nc = pd.DataFrame.from_dict(data_nc, orient='index')
    
    return data_c, data_nc

data_c,data_nc=load_results("gpt_responses.json")
data_merged = pd.merge(data_c, data_nc, on='id', suffixes=('_c', '_nc'))
data_merged['improvement'] = 0  # Default to no change
data_merged.loc[(data_merged['correctness_c'] == True) & (data_merged['correctness_nc'] == False), 'improvement'] = 1
data_merged.loc[(data_merged['correctness_c'] == False) & (data_merged['correctness_nc'] == True), 'improvement'] = -1
# Export data_c DataFrame to a JSON file
data_merged.to_json('data_merged_records.json', orient='records')

```

```python
# Define bins that cover the entire range of 'similarity' values
bins = np.linspace(0.023, 0.128, num=5)
data_merged['similarity_bin'] = pd.cut(data_merged['similarity'], bins, labels=[f"{bins[i]:.2f}-{bins[i+1]:.2f}" for i in range(len(bins)-1)])

# Check how many data points fall into each bin
print(data_merged['similarity_bin'].value_counts())

# Calculating mean and standard deviation per bin
stats_per_bin = data_merged.groupby('similarity_bin')['improvement'].agg(['mean', 'std'])

print(data_merged['similarity_bin'].value_counts())
print(stats_per_bin)
plt.figure(figsize=(12, 6))
# Ensure there are no NaN values; fill with 0 if necessary
stats_per_bin = stats_per_bin.fillna(0)

# Adding error bars; capsize adds horizontal lines at the top of the error bars
plt.bar(stats_per_bin.index, stats_per_bin['mean'], yerr=stats_per_bin['std'], capsize=5, color='blue', alpha=0.7)

plt.title('Improvement Score by Similarity Score Bins')
plt.xlabel('Similarity Score Bins')
plt.ylabel('Mean Improvement Score')
plt.xticks(rotation=45)
plt.ylim([min(stats_per_bin['mean'] - stats_per_bin['std']) - 1, max(stats_per_bin['mean'] + stats_per_bin['std']) + 1])  # Adjust y-axis to show all data
plt.grid(axis='y')
plt.show()
```
