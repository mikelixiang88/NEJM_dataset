```python
import re
import pprint

import hashlib
import json
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML
display(HTML("<style>.container { width:80% !important; }</style>"))
```

```python
def get_similarity_scores_by_id(data_list, target_id):
    for item in data_list:
        if item['ID'] == target_id:
            # Extract and return the two specific values
            return item['question_context_similarity']
    return None     # Return None for both if no match is found

def load_json_data(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    return data

# Load the CSV file
def load_csv(file_path):
    return pd.read_csv(file_path)
    
def filter_answers(df):
    # Use a regular expression to filter rows
    # The regex '^[A-Z]$' matches a single uppercase letter
    return df[df['Answer'].str.match(r'^[A-Z]$', na=False)]

file_path = 'NEJM_All_Questions_And_Answers.csv'  
df = load_csv(file_path)
df = filter_answers(df)

def load_results(filename):
    """Load results."""
    data_c = {}
    embedding_distance = load_json_data('updated_results.json')
    df.set_index('ID', inplace=True)  # Set the index to 'id' for easier lookup
    with open(filename, 'r') as f:
        for line in f:
            v_t = json.loads(line)
            for v in v_t:
                c_len=len(v['context'])
                if c_len==0:
                    continue
                #h = hashlib.md5(v['displayed_text'].encode('utf-8')).hexdigest()
                context_similarity = get_similarity_scores_by_id(embedding_distance,int(v['id']))
                answer=v['response'][0]
                correctness = (answer == df.loc[int(v['id']), 'Answer']) if int(v['id']) in df.index else False
                data_c[v['id']] = {
                    'id': v['id'],
                    'c_lenth':c_len,
                    'correctness':correctness,
                    'similarity':context_similarity,
                }
    
    data_c = pd.DataFrame.from_dict(data_c, orient='index')
    
    return data_c

data_c=load_results("gpt_responses.json")
data_c
```

```python
# Binning the similarity scores
bins = np.linspace(0.02, 0.15, num=10)  # Adjust the number of bins as needed
data_c['similarity_bin'] = pd.cut(data_c['similarity'], bins, labels=[f"{bins[i]:.2f}-{bins[i+1]:.2f}" for i in range(len(bins)-1)])

# Calculating success rate per bin
success_rate = data_c.groupby('similarity_bin')['correctness'].mean()

# Plotting
plt.figure(figsize=(12, 6))
success_rate.plot(kind='bar', color='blue', alpha=0.7)
plt.title('Success Rate by Similarity Score Bins')
plt.xlabel('Similarity Score Bins')
plt.ylabel('Success Rate')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
